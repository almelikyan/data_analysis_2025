## Normality Test

```python
df = pd.read_excel('Worldbank_upd.xlsx')
df = df.dropna(subset=['Travelservicesofserviceex'])
stats.kstest(df.Travelservicesofserviceex, 'norm', args=(df.Travelservicesofserviceex.mean(), df.Travelservicesofserviceex.std()))
```

H1: The distribution of *Travel services (% of service exports)* variable is different from the normal distribution

H0: The distribution of *Travel services (% of service exports)* variable is not different from the normal distribution

Conclusion: as p-value > 0.05, H1 is rejected => the distribution is not significantly different from normal

It means that t-tests can be used to compare mean values of the variable in groups.

```python
sns.distplot(df.Travelservicesofserviceex, fit=stats.norm)  
```

## 2. Independent-samples t-test <a class="anchor" id="chapter2"></a>

```python
from IPython.display import Image
Image("picture1.png")
```

Compare mean values of Travel services (% of service exports) in 2 groups of countries (high income and others).

```python
df.groupby('high_income')['Travelservicesofserviceex'].describe()
```

```python
group1 = df[df['high_income']==0]['Travelservicesofserviceex']
group2 = df[df['high_income']==1]['Travelservicesofserviceex']
```

```python
stats.ttest_ind(group1.dropna(), group2.dropna())
```

H1: Mean values of the *percent of travel services out of service exports* for high income countries and other countries are different.

H0: Mean values of the *percent of travel services out of service exports* for high income countries and other countries are not different.

**Conclusion:** p-value < 0.05, H1 is accepted, H0 is rejected => there is a statistically significant difference in mean values of the dependent variable in the two groups

### Levene's Test for Equality of Variances 

Levene’s test is a statistical procedure used to assess whether the variances of a variable are equal across two or more groups.

```python
stats.levene(group1.dropna(), group2.dropna())
```

H1: Variances of the *percent of travel services out of service exports* for high income countries and other countries are different.

H0: Variances of the *percent of travel services out of service exports* for high income countries and other countries are not different.

**Conclusion:** p-value > 0.05, H1 is rejected, H0 is not rejected => variances in two groups are not different

If variances are different its recommended to use a t-test with Welch correction. In this case the degrees of freedom are calculated differently, they are less than in the standard t-test calculation.

```python
stats.ttest_ind(group1.dropna(), group2.dropna(), equal_var=False)
```

## 5. ANOVA <a class="anchor" id="chapter5"></a>

```python
Image("anova6.png")
```

```python
df = pd.read_excel('Worldbank.xlsx')
group_high = df[df['income']=='high income']['Travelservicesofserviceex']
group_low = df[df['income']=='low income']['Travelservicesofserviceex']
group_middle = df[df['income']=='middle income']['Travelservicesofserviceex']
```

```python
Image("anova3.png")
```

```python
stats.f_oneway(group_high, group_low, group_middle)
```

H1: the mean values of the *percent of travel services out of service exports* for groups of countries based on income are different

H0: the mean values of the *percent of travel services out of service exports* for groups of countries based on income are not different

**Conclusion:** p-value < 0.05, H1 is accepted, H0 is rejected => means are different.

Pairwise post-hoc testing will define whether there is a difference between the means of all possible pairs (all pairwise comparisons).

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
print(pairwise_tukeyhsd(df['Travelservicesofserviceex'], df['income']))
```

**Conclusion:** according to the results of the post-hoc test, statistically significant differences in the mean values of *Travel services (% of service exports)* are observed between high income and middle income groups, as well as between low income and middle income groups.

## 8. Kruskal-Wallis test <a class="anchor" id="chapter8"></a>

```python
Image("picture7.png")
```

```python
group_high = df[df['income']=='high income']['Agriculturalrawmaterialsexpor']
group_low = df[df['income']=='low income']['Agriculturalrawmaterialsexpor']
group_middle = df[df['income']=='middle income']['Agriculturalrawmaterialsexpor']
```

```python
stats.kruskal(group_high, group_low, group_middle)
```

H1: Values of the variable *Agricultural raw materials exports (% of merchandise exports)* for groups of countries based on income are different.

H0: Values of the variable *Agricultural raw materials exports (% of merchandise exports)* for groups of countries based on income are not different.

**Conclusion:** p-value < 0.05, H1 is accepted, H0 is rejected, values are different.

To find out in which groups the values differ, we can conduct Mann-Whitney tests between all possible pairs of groups.

```python
from itertools import combinations
list_groups = ['group_high', 'group_low', 'group_middle']
for a, b in combinations(list_groups, 2):
    print(f'{a} and {b}:', stats.mannwhitneyu(eval(a), eval(b))[1])
```

**Conclusion:** According to the results, statistically significant differences in the values of the variable *Agricultural raw materials exports (% of merchandise exports)* are observed between high income and middle income groups, as well as between low income and middle income groups.
